```python
genes = [
"MMP9","LINC00152","ESM1","CLIC1","SERPINH1","AGAP2-AS1","KCNB1",
"ANXA2","LOX","CCDC109B","EMP3","PLEK2","SH2D4A","TUBA1C",
"FN1","TNFRSF12A","RAB34","CA9","TAGLN2","GDF15","SPAG4"
]
```

```python
import gseapy as gp

go_results = gp.enrichr(
    gene_list=genes,
    gene_sets="GO_Biological_Process_2021",
    organism="human",
    outdir=None
)

go_results.results.head(10)
```

```python
kegg_results = gp.enrichr(
    gene_list=genes,
    gene_sets="KEGG_2021_Human",
    organism="human",
    outdir=None
)

kegg_results.results.head(10)
```

```python
import matplotlib.pyplot as plt

gp.barplot(
    go_results.results,
    title="GO Biological Process Enrichment",
    top_term=10
)
plt.show()
```

```python
reactome_results = gp.enrichr(
    gene_list=genes,
    gene_sets="Reactome_2022",
    organism="human",
    outdir=None
)

reactome_results.results.head(15)
```

```python
reactome_results.results.head(15)
```

```python
gp.barplot(
    reactome_results.results,
    title="Reactome Pathway Enrichment",
    top_term=10,
    cutoff=0.05
)
```

Reactome pathway enrichment analysis revealed significant involvement of extracellular matrix organization and collagen formation pathways. These pathways are closely associated with tumor invasion and microenvironment remodeling in glioma, suggesting that the machine learning–identified biomarkers may contribute to glioma progression through extracellular matrix regulation.
