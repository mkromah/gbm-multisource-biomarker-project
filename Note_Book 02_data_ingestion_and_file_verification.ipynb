## Notebook 02 — Data Ingestion and File Verification

This notebook verifies that all raw datasets required for the glioma machine learning project are present and accessible from Google Drive.




**Mount Drive and load shared configuration**

```python
# ============================================================
# Step 1: Mount Google Drive and define shared project paths
# ============================================================

from google.colab import drive
drive.mount('/content/drive')

import os
from pathlib import Path
import pandas as pd

PROJECT_NAME = "Glioma_ML_Biomarker_Project"
BASE_DIR = f"/content/drive/MyDrive/{PROJECT_NAME}"

DATA_DIR = f"{BASE_DIR}/data"
RAW_DIR = f"{DATA_DIR}/raw"
PROCESSED_DIR = f"{DATA_DIR}/processed"

CGGA_DIR = f"{RAW_DIR}/CGGA"
XENA_DIR = f"{RAW_DIR}/XENA"
GEO_DIR = f"{RAW_DIR}/GEO"

RESULTS_DIR = f"{BASE_DIR}/results"
FIGURES_DIR = f"{RESULTS_DIR}/figures"
TABLES_DIR = f"{RESULTS_DIR}/tables"
MODELS_DIR = f"{BASE_DIR}/models"

print("Shared project configuration loaded successfully.")
```

```python
# ============================================================
# Step 2: Verify that the main project directories exist
# ============================================================

project_directories = {
    "BASE_DIR": BASE_DIR,
    "DATA_DIR": DATA_DIR,
    "RAW_DIR": RAW_DIR,
    "CGGA_DIR": CGGA_DIR,
    "XENA_DIR": XENA_DIR,
    "GEO_DIR": GEO_DIR,
    "PROCESSED_DIR": PROCESSED_DIR,
    "RESULTS_DIR": RESULTS_DIR,
    "FIGURES_DIR": FIGURES_DIR,
    "TABLES_DIR": TABLES_DIR,
    "MODELS_DIR": MODELS_DIR
}

for name, path in project_directories.items():
    status = "Exists" if os.path.exists(path) else "Missing"
    print(f"{name:<15} | {status:<7} | {path}")
```

```python
# ============================================================
# Step 2 — Verify raw-data folders


import os

raw_folders = {
    "CGGA_DIR": CGGA_DIR,
    "XENA_DIR": XENA_DIR,
    "GEO_DIR": GEO_DIR
}

for folder_name, folder_path in raw_folders.items():
    exists = os.path.exists(folder_path)
    status = "FOUND" if exists else "MISSING"
    print(f"{folder_name}: {status} -> {folder_path}")
```

```python
# ============================================================
# Step 3: Define a helper function to inspect folder contents
# ============================================================

def inspect_folder(folder_path):
    folder = Path(folder_path)

    if not folder.exists():
        print(f"[Missing folder] {folder_path}")
        return pd.DataFrame()

    items = list(folder.iterdir())

    if len(items) == 0:
        print(f"[Empty folder] {folder_path}")
        return pd.DataFrame()

    records = []

    for item in sorted(items):
        item_type = "Folder" if item.is_dir() else "File"
        size_mb = round(item.stat().st_size / (1024 ** 2), 2) if item.is_file() else None

        records.append({
            "name": item.name,
            "type": item_type,
            "size_mb": size_mb,
            "path": str(item)
        })

    df = pd.DataFrame(records)
    print(f"\nFolder inspected: {folder_path}")
    print(f"Total items found: {len(df)}")
    return df
```

```python
# ============================================================
# Step 4: Inspect the CGGA raw-data folder
# ============================================================

cgga_files = inspect_folder(CGGA_DIR)
cgga_files
```

```python
# ============================================================
# Step 5: Inspect the XENA raw-data folder
# ============================================================

xena_files = inspect_folder(XENA_DIR)
xena_files
```

```python
# ============================================================
# Step 6: Inspect the GEO raw-data folder

geo_files = inspect_folder(GEO_DIR)
geo_files
```

```python
# ============================================================
# Step 7: Combine raw-data inventories across all sources
# ============================================================

def add_source_column(df, source_name):
    if df.empty:
        return df
    df = df.copy()
    df.insert(0, "source", source_name)
    return df

inventory_frames = [
    add_source_column(cgga_files, "CGGA"),
    add_source_column(xena_files, "XENA"),
    add_source_column(geo_files, "GEO")
]

inventory_frames = [df for df in inventory_frames if not df.empty]

raw_inventory_df = pd.concat(inventory_frames, ignore_index=True)

print("Combined raw-data inventory created successfully.")
raw_inventory_df
```

```python
# ============================================================
# Step 8: Save the raw-data inventory table
# ============================================================

TABLES_DIR = f"{BASE_DIR}/results/tables"
os.makedirs(TABLES_DIR, exist_ok=True)

inventory_path = f"{TABLES_DIR}/raw_data_inventory.csv"

raw_inventory_df.to_csv(inventory_path, index=False)

print("Raw data inventory saved successfully.")
print("Location:", inventory_path)
```
