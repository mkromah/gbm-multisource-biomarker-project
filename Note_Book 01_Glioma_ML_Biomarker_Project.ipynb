```python
# ============================================================
# Step 1: Mount Google Drive
# ============================================================
# Why this step exists:
# Google Colab runs in a temporary cloud environment.
# Mounting Google Drive allows us to store project files
# permanently and reuse them across sessions.
#
# Expected outcome:
# A Drive authorization prompt appears, and after approval,
# Google Drive becomes available under /content/drive
# ============================================================

from google.colab import drive
drive.mount('/content/drive')
```

```python
# ============================================================
# Step 2: Defining project name and root directory
# ============================================================
# This block sets the official internal project name and
# determines where the project will live inside Google Drive.
# ============================================================

import os

PROJECT_NAME = "Glioma_ML_Biomarker_Project"
BASE_DIR = f"/content/drive/MyDrive/{PROJECT_NAME}"

print("Project name:", PROJECT_NAME)
print("Project root directory:", BASE_DIR)
```

```python
# ============================================================
# Step 3: Creating the project folder structure
# ============================================================


folders = [
    BASE_DIR,
    f"{BASE_DIR}/data",
    f"{BASE_DIR}/data/raw",
    f"{BASE_DIR}/data/raw/CGGA",
    f"{BASE_DIR}/data/raw/XENA",
    f"{BASE_DIR}/data/raw/GEO",
    f"{BASE_DIR}/data/processed",
    f"{BASE_DIR}/results",
    f"{BASE_DIR}/results/figures",
    f"{BASE_DIR}/results/tables",
    f"{BASE_DIR}/models",
    f"{BASE_DIR}/notebooks"
]

for folder in folders:
    os.makedirs(folder, exist_ok=True)

print("Project folders created successfully.")
```

```python
# ============================================================
# Step 4: Verify that all required folders exist
# ============================================================

for folder in folders:
    status = "Exists" if os.path.exists(folder) else "Missing"
    print(f"{status:>7} | {folder}")
```

```python
# ============================================================
# Step 6: Import core libraries
# ============================================================
# These imports are grouped by purpose:
# - system/path handling
# - data handling
# - visualization
# - machine learning
# ============================================================

import os
import gzip
import shutil
import warnings

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    roc_auc_score,
    confusion_matrix
)
from sklearn.ensemble import RandomForestClassifier

warnings.filterwarnings("ignore")

print("Core libraries imported successfully.")
```

```python
# ============================================================
# Step 7: Configure display and plotting options
# ============================================================
# These settings make notebook outputs easier to read.
# They do not change the data itself.
# ============================================================

pd.set_option("display.max_columns", 100)
pd.set_option("display.max_rows", 30)
pd.set_option("display.width", 150)
pd.set_option("display.max_colwidth", 120)

plt.rcParams["figure.figsize"] = (10, 6)

print("Display settings configured.")
```

```python
# ============================================================
# Step 8: Define reusable path variables
# ============================================================
# These variables act as a central reference for all project
# locations. Later notebooks should import or redefine these
# exactly the same way.
# ============================================================

DATA_DIR = f"{BASE_DIR}/data"
RAW_DIR = f"{DATA_DIR}/raw"
PROCESSED_DIR = f"{DATA_DIR}/processed"

CGGA_DIR = f"{RAW_DIR}/CGGA"
XENA_DIR = f"{RAW_DIR}/XENA"
GEO_DIR = f"{RAW_DIR}/GEO"

RESULTS_DIR = f"{BASE_DIR}/results"
FIGURES_DIR = f"{RESULTS_DIR}/figures"
TABLES_DIR = f"{RESULTS_DIR}/tables"

MODELS_DIR = f"{BASE_DIR}/models"
NOTEBOOKS_DIR = f"{BASE_DIR}/notebooks"

print("Path variables defined successfully.")
```

```python
# ============================================================
# Step 9: Print a human-readable summary of important paths
# ============================================================

project_paths = {
    "BASE_DIR": BASE_DIR,
    "DATA_DIR": DATA_DIR,
    "RAW_DIR": RAW_DIR,
    "CGGA_DIR": CGGA_DIR,
    "XENA_DIR": XENA_DIR,
    "GEO_DIR": GEO_DIR,
    "PROCESSED_DIR": PROCESSED_DIR,
    "RESULTS_DIR": RESULTS_DIR,
    "FIGURES_DIR": FIGURES_DIR,
    "TABLES_DIR": TABLES_DIR,
    "MODELS_DIR": MODELS_DIR,
    "NOTEBOOKS_DIR": NOTEBOOKS_DIR
}

for name, path in project_paths.items():
    print(f"{name:<15} -> {path}")
```

```python
# ============================================================
# Step 10: Final environment readiness check
# ============================================================

all_ok = True

for name, path in project_paths.items():
    exists = os.path.exists(path)
    print(f"{name:<15} | {'Exists' if exists else 'Missing'}")
    if not exists:
        all_ok = False

if all_ok:
    print("\nEnvironment setup completed successfully.")
else:
    print("\nSome required paths are missing. Please review the setup steps.")
```
