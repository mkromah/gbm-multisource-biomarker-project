```python
# ============================================================
# Step 2: Define shared project paths
# ============================================================

import os
import gzip
import pandas as pd

PROJECT_NAME = "Glioma_ML_Biomarker_Project"
BASE_DIR = f"/content/drive/MyDrive/{PROJECT_NAME}"

DATA_DIR = f"{BASE_DIR}/data"
RAW_DIR = f"{DATA_DIR}/raw"
PROCESSED_DIR = f"{DATA_DIR}/processed"

XENA_DIR = f"{RAW_DIR}/XENA"

print("Project root:", BASE_DIR)
print("XENA raw-data folder:", XENA_DIR)
```

```python
# ============================================================
# Step 3: Define XENA file paths
# ============================================================

XENA_PHENOTYPE_FILE = f"{XENA_DIR}/TcgaTargetGTEX_phenotype.txt.gz"
XENA_EXPRESSION_FILE = f"{XENA_DIR}/TcgaTargetGtex_rsem_gene_tpm.gz"
XENA_PROBEMAP_FILE = f"{XENA_DIR}/probeMap_gencode.v23.annotation.gene.probemap"

print("Phenotype file:", XENA_PHENOTYPE_FILE)
print("Expression file:", XENA_EXPRESSION_FILE)
print("Probe-map file:", XENA_PROBEMAP_FILE)
```

```python
# ============================================================
# Step 4: Verify that the XENA files exist
# ============================================================

xena_file_paths = {
    "XENA_PHENOTYPE_FILE": XENA_PHENOTYPE_FILE,
    "XENA_EXPRESSION_FILE": XENA_EXPRESSION_FILE,
    "XENA_PROBEMAP_FILE": XENA_PROBEMAP_FILE
}

for file_name, file_path in xena_file_paths.items():
    exists = os.path.exists(file_path)
    status = "FOUND" if exists else "MISSING"
    print(f"{file_name}: {status} -> {file_path}")
```

```python
# ============================================================
# Step 5: Load the XENA phenotype metadata safely
# ============================================================
xena_pheno_df = pd.read_csv(
    XENA_PHENOTYPE_FILE,
    sep="\t",
    compression="gzip",
    encoding="latin1"
)

print("XENA phenotype metadata loaded successfully.")
print("Shape:", xena_pheno_df.shape)
```

```python
print("First 5 rows of the phenotype table:")
display(xena_pheno_df.head())

print("\nColumn names:")
print(list(xena_pheno_df.columns))
```

```python
# ============================================================
# Step 7: List all disease categories
# ============================================================

disease_counts = xena_pheno_df["detailed_category"].value_counts()

print("Number of unique disease categories:", disease_counts.shape[0])

disease_counts.head(20)
```

```python
# ============================================================
# Step 8: Check for Glioblastoma samples safely
# ============================================================

gbm_preview = xena_pheno_df[
    xena_pheno_df["detailed_category"].str.contains("Glioblastoma", case=False, na=False)
]

print("Number of rows matching 'Glioblastoma':", gbm_preview.shape[0])
display(gbm_preview.head())
```

```python
# ============================================================
# Step 9: List glioma-related disease labels
# ============================================================

glioma_like = xena_pheno_df[
    xena_pheno_df["detailed_category"].str.contains("Glioma|Glioblastoma", case=False, na=False)
]["detailed_category"].value_counts()

print("Glioma-related categories found:")
display(glioma_like)
```

```python
# ============================================================
# Step 10: Extract glioma cohort from phenotype table
# ============================================================

glioma_categories = [
    "Brain Lower Grade Glioma",
    "Glioblastoma Multiforme"
]

glioma_pheno_df = xena_pheno_df[
    xena_pheno_df["detailed_category"].isin(glioma_categories)
].copy()

print("Glioma cohort extracted.")
print("Number of samples:", glioma_pheno_df.shape[0])

glioma_pheno_df.head()
```

```python
# ============================================================
# Step 11: Keep only primary tumor samples
# ============================================================

glioma_tumor_df = glioma_pheno_df[
    glioma_pheno_df["_sample_type"] == "Primary Tumor"
].copy()

print("Primary tumor samples retained.")
print("Number of tumor samples:", glioma_tumor_df.shape[0])

glioma_tumor_df.head()
```

```python
# ============================================================
# Step 12: Extract sample IDs for the glioma tumor cohort
# ============================================================

glioma_sample_ids = glioma_tumor_df["sample"].tolist()

print("Number of glioma tumor samples:", len(glioma_sample_ids))

print("\nFirst 10 sample IDs:")
print(glioma_sample_ids[:10])
```

```python
# ============================================================
# Step 13: Inspect expression matrix header
# ============================================================

import gzip

with gzip.open(XENA_EXPRESSION_FILE, "rt") as f:
    header_line = f.readline()

expression_columns = header_line.strip().split("\t")

print("Number of columns in expression matrix:", len(expression_columns))

print("\nFirst 20 columns:")
print(expression_columns[:20])
```

```python
glioma_sample_ids
```

```python
# ============================================================
# Step 14: Identify expression columns for glioma samples
# ============================================================

expression_sample_ids = expression_columns[1:]  # skip gene column

glioma_expression_columns = [
    s for s in expression_sample_ids if s in glioma_sample_ids
]

print("Total expression samples:", len(expression_sample_ids))
print("Matched glioma samples:", len(glioma_expression_columns))

print("\nFirst 10 matched samples:")
print(glioma_expression_columns[:10])
```

```python
# ============================================================
# Step 15: Extract glioma expression data from XENA matrix
# ============================================================

# Columns to load
columns_to_load = ["sample"] + glioma_expression_columns

# Load only required columns
xena_glioma_expr_df = pd.read_csv(
    XENA_EXPRESSION_FILE,
    sep="\t",
    compression="gzip",
    usecols=columns_to_load
)

print("XENA glioma expression matrix loaded successfully.")
print("Matrix shape:", xena_glioma_expr_df.shape)
```

```python
# ============================================================
# Step 16: Inspect extracted expression matrix
# ============================================================

print("First 5 rows of the matrix:")
display(xena_glioma_expr_df.head())

print("\nFirst 10 column names:")
print(list(xena_glioma_expr_df.columns[:10]))
```

```python
# ============================================================
# Step 17: Transpose expression matrix
# ============================================================

# Set gene column as index
xena_glioma_expr_df = xena_glioma_expr_df.set_index("sample")

# Transpose matrix
xena_glioma_expr_t = xena_glioma_expr_df.T

print("Matrix transposed successfully.")

print("New shape:", xena_glioma_expr_t.shape)
```

```python
# ============================================================
# Step 18: Attach tumor labels
# ============================================================

# Extract sample labels from phenotype table
sample_labels = glioma_tumor_df[["sample", "detailed_category"]]

# Set index for merging
sample_labels = sample_labels.set_index("sample")

# Add label column to expression matrix
xena_glioma_expr_t["tumor_label"] = sample_labels.loc[
    xena_glioma_expr_t.index,
    "detailed_category"
]

print("Tumor labels added successfully.")

# Check distribution
print("\nLabel distribution:")
print(xena_glioma_expr_t["tumor_label"].value_counts())
```

```python
import numpy as np
```

```python
# ============================================================
# Step 19: Replace XENA missing value indicator
# ============================================================

XENA_MISSING_VALUE = -9.9658

xena_glioma_expr_t = xena_glioma_expr_t.replace(XENA_MISSING_VALUE, np.nan)

print("Missing value indicator replaced with NaN.")
```

```python
# ============================================================
# Step 20: Check missing values
# ============================================================

missing_counts = xena_glioma_expr_t.isna().sum().sum()

print("Total missing values in dataset:", missing_counts)
```

```python
# ============================================================
# Step 21: Remove genes with too many missing values
# ============================================================

# Separate labels from expression matrix
labels = xena_glioma_expr_t["tumor_label"]
expr = xena_glioma_expr_t.drop(columns=["tumor_label"])

# Calculate allowed missing threshold
missing_threshold = int(expr.shape[0] * 0.2)

# Filter genes
expr_filtered = expr.dropna(axis=1, thresh=expr.shape[0] - missing_threshold)

print("Original gene count:", expr.shape[1])
print("Filtered gene count:", expr_filtered.shape[1])
```

```python
# ============================================================
# Step 22: Reattach tumor labels
# ============================================================

expr_filtered = expr_filtered.copy()
expr_filtered.loc[:, "tumor_label"] = labels

print("Dataset ready after missing-value filtering.")
print("Final dataset shape:", expr_filtered.shape)
```

```python
# ============================================================
# Step 23: Remove low-variance genes
# ============================================================

from sklearn.feature_selection import VarianceThreshold

# Separate features and labels
y = expr_filtered["tumor_label"]
X = expr_filtered.drop(columns=["tumor_label"])

# Apply variance filtering
variance_filter = VarianceThreshold(threshold=0.01)

X_var_array = variance_filter.fit_transform(X)

# Recover selected gene names
selected_genes = X.columns[variance_filter.get_support()]

# Rebuild as DataFrame
X_var = pd.DataFrame(
    X_var_array,
    index=X.index,
    columns=selected_genes
)

print("Genes before variance filtering:", X.shape[1])
print("Genes after variance filtering:", X_var.shape[1])
print("Filtered feature matrix shape:", X_var.shape)
```

```python
# ============================================================
# Step 24: Select top variable genes
# ============================================================

# Calculate variance for each gene
gene_variances = X.var(axis=0)

# Sort genes by variance
top_genes = gene_variances.sort_values(ascending=False).head(5000).index

# Subset dataset
X_var = X[top_genes]

print("Original gene count:", X.shape[1])
print("Selected high-variance genes:", X_var.shape[1])
print("New feature matrix shape:", X_var.shape)
```

```python
# ============================================================
# Step 25: Reattach labels safely
# ============================================================

X_var = X_var.copy()
X_var.loc[:, "tumor_label"] = y

print("Final ML dataset shape:", X_var.shape)
```

```python
# ============================================================
# Step 26: Encode tumor labels
# ============================================================

label_mapping = {
    "Brain Lower Grade Glioma": 0,
    "Glioblastoma Multiforme": 1
}

# Separate features and labels
X_ml = X_var.drop(columns=["tumor_label"]).copy()
y_ml = X_var["tumor_label"].map(label_mapping)

print("Label encoding completed successfully.")

print("\nEncoded label distribution:")
print(y_ml.value_counts().sort_index())
```

```python
# ============================================================
# Step 27: Train-test split
# ============================================================

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_ml,
    y_ml,
    test_size=0.2,
    random_state=42,
    stratify=y_ml
)

print("Train-test split completed successfully.")

print("\nTraining set shape:", X_train.shape)
print("Test set shape:", X_test.shape)

print("\nTraining label distribution:")
print(y_train.value_counts().sort_index())

print("\nTest label distribution:")
print(y_test.value_counts().sort_index())
```

```python
# ============================================================
# Step 28: Train baseline Random Forest model
# ============================================================

from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    n_jobs=-1
)

rf_model.fit(X_train, y_train)

print("Random Forest model trained successfully.")
```

```python
# ============================================================
# Step 29: Evaluate baseline Random Forest model
# ============================================================

from sklearn.metrics import accuracy_score, classification_report, roc_auc_score

# Predictions
y_pred = rf_model.predict(X_test)
y_prob = rf_model.predict_proba(X_test)[:, 1]

# Metrics
accuracy = accuracy_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_prob)

print("Baseline Random Forest Performance")
print("-" * 40)

print("Accuracy:", round(accuracy, 4))
print("ROC-AUC :", round(roc_auc, 4))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
# ============================================================
# Step 30: Extract gene importance scores
# ============================================================

import pandas as pd

# Extract importance scores
importance_scores = rf_model.feature_importances_

# Create dataframe
importance_df = pd.DataFrame({
    "gene": X_train.columns,
    "importance": importance_scores
})

# Sort genes by importance
importance_df = importance_df.sort_values(
    by="importance",
    ascending=False
)

print("Top 10 most important genes:")
display(importance_df.head(10))
```

```python
# ============================================================
# Step 31: Save ranked biomarker genes
# ============================================================

import os

output_path = "/content/drive/MyDrive/Glioma_ML_Biomarker_Project/results"

os.makedirs(output_path, exist_ok=True)

importance_df.to_csv(
    f"{output_path}/xena_gene_importance_ranking.csv",
    index=False
)

print("Gene importance ranking saved successfully.")
```

```python
# ============================================================
# Step 32: Load gene annotation map (Ensembl → Gene Symbol)
# ============================================================

PROBEMAP_FILE = "/content/drive/MyDrive/Glioma_ML_Biomarker_Project/data/raw/XENA/probeMap_gencode.v23.annotation.gene.probemap"

gene_map = pd.read_csv(
    PROBEMAP_FILE,
    sep="\t"
)

print("Gene annotation map loaded successfully.")
print("Shape:", gene_map.shape)

gene_map.head()
```

```python
# ============================================================
# Step 33: Map Ensembl IDs to gene symbols
# ============================================================

# Remove version numbers from Ensembl IDs (e.g., ENSG00000223972.5 → ENSG00000223972)
importance_df["ensembl_id"] = importance_df["gene"].str.split(".").str[0]
gene_map["ensembl_id"] = gene_map["id"].str.split(".").str[0]

# Merge importance table with gene symbol mapping
importance_annotated = importance_df.merge(
    gene_map[["ensembl_id", "gene"]],
    on="ensembl_id",
    how="left"
)

# Rename columns for clarity
importance_annotated = importance_annotated.rename(
    columns={
        "gene_x": "ensembl_gene",
        "gene_y": "gene_symbol"
    }
)

print("Gene annotation completed.")

importance_annotated.head(10)
```

```python
# ============================================================
# Step 34: Extract top biomarker genes
# ============================================================

TOP_N = 20

top_biomarkers = importance_annotated.head(TOP_N)

print(f"Top {TOP_N} candidate biomarkers:")
display(top_biomarkers)
```
